```python
import pandas as pd

# Baca file CSV
file_path = r'C:\Project\ML_learn\Sample\Akses 750+ Produk Digital by RKV - List Product Digital.csv'
df = pd.read_csv(file_path, encoding='utf-8')

# Hapus kolom 'No' karena mengganggu pengecekan duplikat
df = df.drop(columns=['No'])

# Tampilkan info awal
print(f"Kolom yang tersisa: {df.columns.tolist()}")
print(f"Total baris sebelum pembersihan: {len(df)}")
print(f"Total duplikat: {df.duplicated().sum()}")

# Hapus duplikat
df_clean = df.drop_duplicates()

# Tampilkan info setelah pembersihan
print(f"Total baris setelah pembersihan: {len(df_clean)}")

# Kembalikan kolom 'No' dengan nomor urut baru (1, 2, 3, ...)
df_clean = df_clean.reset_index(drop=True)
df_clean.insert(0, 'No', df_clean.index + 1)

# Simpan hasil ke file Excel
output_path = r'C:\Project\ML_learn\Sample\Akses List Product Digital.xlsx'
df_clean.to_excel(output_path, index=False, engine='openpyxl')

print(f"\nFile tersimpan di: {output_path}")
print(f"\nPreview data:")
df_clean.head(10)
```
